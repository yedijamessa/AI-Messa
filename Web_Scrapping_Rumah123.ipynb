```python
# URL of the website (you need to replace with the actual URL)
url = 'https://www.lamudi.co.id/newdevelopments/lamudi-projects-agn/grand-tenjo-residence-1/'

# Set up the headers with a User-Agent
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

# Make the request with the headers
response = requests.get(url, headers=headers)
```

```python
# If the status code is 200, the request was successful
if response.status_code == 200:

    # Parse the HTML content
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all the 'li' elements with class 'Card' containing the price info
    card_items = soup.find_all('li', class_='Card')

    # Loop through each 'li' element to find the price within a 'span' with class 'value'
    for card in card_items:
        price_container = card.find('div', class_='Card-info-pricetype')

        # Check if the price container is found
        if price_container:
            price_value = price_container.find('span', class_='value')

            # If the price value is found, print it
            if price_value:
                print(price_value.text.strip())
            else:
                print("Price value not found in the container.")
        else:
            print("Price container not found.")
else:
    print(f"Failed to retrieve the webpage, status code: {response.status_code}")
```

```python
import requests
from bs4 import BeautifulSoup

# Base URL of the website without the page number
base_url = 'https://www.rumah123.com/jual/yogyakarta/rumah/?minPrice=null&maxPrice=null'

# Set up the headers with a User-Agent to mimic a browser
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

# Define the number of pages you want to scrape
number_of_pages = 10  # for example, if you want to scrape 10 pages

# Loop through the specified range of pages
for page in range(1, number_of_pages + 1):
    # Append the page number to the base URL
    url = f"{base_url}&page={page}"
    print(f"Scraping {url}")

    # Make the request with the headers
    response = requests.get(url, headers=headers)

    # Check if the request was successful
    if response.status_code == 200:
        # Create a BeautifulSoup object and specify the parser
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find all elements with the specified class
        price_elements = soup.find_all('div', class_='card-featured__middle-section__price')

        # Loop through all found elements and print their text content
        for price_section in price_elements:
            price = price_section.find('strong')
            if price:
                print(price.text.strip())  # .strip() removes leading/trailing whitespace
            else:
                print("Price element not found.")
    else:
        print(f"Failed to retrieve the webpage, status code: {response.status_code}")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Function to get the total number of pages
def get_total_pages(url):
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        pagination_container = soup.find('ul', class_='ui-molecule-paginate')
        if pagination_container:
            page_links = pagination_container.find_all('a', href=True)
            last_page_link = page_links[-2]['href']
            number_of_pages = int(last_page_link.split('page=')[1])
            return number_of_pages
    return 0

# Function to scrape prices from a single page
def scrape_prices_from_page(url):
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        price_elements = soup.find_all('div', class_='card-featured__middle-section__price')
        prices = [price.find('strong').text.strip() for price in price_elements if price.find('strong')]
        return prices
    return []

# The base URL for the first page of listings
base_url = 'https://www.rumah123.com/jual/yogyakarta/rumah/?minPrice=null&maxPrice=null'

# Set up the headers with a User-Agent to mimic a browser
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

# Get the total number of pages from the website
total_pages = get_total_pages(base_url)

# List to store all prices
all_prices = []

# Loop through all pages and scrape the prices
for page in range(1, total_pages + 1):
    page_url = f"{base_url}&page={page}"
    prices_on_page = scrape_prices_from_page(page_url)
    all_prices.extend(prices_on_page)

# Create a DataFrame with the prices
df_prices = pd.DataFrame(all_prices, columns=['Price'])

# # Output the DataFrame
# print(df_prices)

# # Optionally, save the DataFrame to a CSV file
# df_prices.to_csv('property_prices.csv', index=False)
```
