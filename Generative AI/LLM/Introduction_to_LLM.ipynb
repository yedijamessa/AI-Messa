**Summarization Using Hugging Face Models**

```python
from transformers import pipeline

summarizer = pipeline(task="summarization", model="facebook/bart-large-cnn")

text = "Walking amid Gion's Machiya wooden houses is a mesmerizing experience. ..."
summary = summarizer(text, max_length=50)
print(summary[0]["summary_text"])
```

* This code demonstrates how to use a pre-trained summarization model (facebook/bart-large-cnn) from Hugging Face's transformers library.
* It takes a piece of text and generates a summarized version with a specified maximum length.

**Text Generation**



```python
from transformers import pipeline

generator = pipeline(task="text-generation", model="distilgpt2")

prompt = "The Gion neighborhood in Kyoto is famous for"
output = generator(prompt, max_length=100, pad_token_id=generator.tokenizer.eos_token_id)
print(output[0]["generated_text"])
```

* This example demonstrates how to guide the text generation process by providing a structured prompt.
* It simulates a conversation by combining a book review and a response in the prompt.

**Language Translation**

```python
from transformers import pipeline

translator = pipeline(task="translation_en_to_es", model="Helsinki-NLP/opus-mt-en-es")

text = "Walking amid Gion's Machiya wooden houses was a mesmerizing experience."
output = translator(text, clean_up_tokenization_spaces=True)
print(output[0]["translation_text"])
```

* This code translates English text into Spanish using the Helsinki-NLP/opus-mt-en-es model.
* The clean_up_tokenization_spaces parameter removes unwanted spaces in the output.

** Transformer Architectures: Encoder-Only**

```python
# Import the pipeline function
from transformers import pipeline

# Create a pipeline using an encoder-only model (bert-base-uncased)
llm = pipeline(model="bert-base-uncased")

# Print the model details
print(llm.model)

# Print the model configuration
print(llm.model.config)

# Check if the model is decoder-based or encoder-based
print(llm.model.config.is_decoder)  # Output: False (indicating encoder-only)
```

* Encoder-only models, like BERT, are typically used for tasks like text classification or sentiment analysis.
* This code inspects the model and its configuration to confirm it is encoder-only.


**Transformer Architectures: Decoder-Only
python**

```python
# Import the pipeline function
from transformers import pipeline

# Create a pipeline using a decoder-only model (gpt2)
llm = pipeline(model="gpt2")

# Print the model configuration
print(llm.model.config)

# Check if the model is decoder-based
print(llm.model.config.is_decoder)  # Output: True (indicating decoder-only)
```

* Decoder-only models, like GPT-2, are used for tasks like text generation or generative question-answering.
* This code confirms the model type by inspecting the configuration.

**Transformer Architectures: Encoder-Decoder**

```python
# Import the pipeline function
from transformers import pipeline

# Create a pipeline using an encoder-decoder model (Helsinki-NLP/opus-mt-es-en)
llm = pipeline(model="Helsinki-NLP/opus-mt-es-en")

# Print the model details
print(llm.model)

# Print the model configuration
print(llm.model.config)

# Check if the model is encoder-decoder
print(llm.model.config.is_encoder_decoder)  # Output: True
```

* Encoder-decoder models are used for tasks like translation or summarization.
* This example demonstrates how to inspect and confirm the architecture of such a model.
